### Add in query info

```python
df = cands
#1. get information on number of queries (standardize this by candidate popularity??)
#2. average num of gendered words
#3. max num of gendered words
#4. min number of gendered words
#5. average length of query
#6. num times query w most gendered language showed up? /popularity
df['num_queries'] = ''
df['num_queries_unique'] = ''
df['num_gen'] = ''
df['avg_len'] = ''
df['avg_num_gen'] = ''
#df['min_num_gen'] = ''
#df['max_num_gen'] = ''
```

```python
qs = pd.read_csv('all_queries_with_data.csv',index_col=False)
qs = qs.fillna("")
qs = qs.replace('annisa','annissa',regex=True)
```

```python
qs[qs.city!='Boston']
```

Add count of gendered words to the query frame

```python
#gendered lexicon
gen_set = ['wife','married','daughter','husband','partner','family','kids','love','son','father','mother','children',
          'boy','girl','marriage']
```

```python
def add_gender_language(query):
    count = 0
    for word in gen_set:
        if word in query.lower(): 
            count = count+1
    return count

qs['gen_count'] = qs.Query.apply(add_gender_language)
```

```python
def get_cand_queries(candidate):
    '''helper function that takes a candidate name and collects all queries searched about that candidate'''
    a = qs[qs.candidate==candidate]
    return a.Query.to_list()
```

```python
def get_avg_gen_count(candidate):
    '''helper function that calculates the average number of gendered words in a query for a candidates'''
    a = qs[qs.candidate==candidate]
    return statistics.mean(a.gen_count.to_list())
```

```python
def get_gen_count(candidate):
    '''helper function that calculates the percentage of candidate's queries that have at least one gendered word'''
    a = qs[(qs.candidate==candidate)&(qs.gen_count>0)]
    return len(a.gen_count.to_list())
```

```python
df['num_queries'] = df['name'].apply(lambda x: len(get_cand_queries(x)))
df['num_queries_unique'] = df['name'].apply(lambda x: len(set(get_cand_queries(x))))
df['num_gen'] = df['name'].apply(lambda x: get_gen_count(x))
df['avg_len'] = df['name'].apply(lambda x: statistics.mean([len(query) for query in get_cand_queries(x)]))
df['avg_num_gen'] = df['name'].apply(lambda x: get_avg_gen_count(x))
#df['min_num_gen'] = ''
#df['max_num_gen'] = ''
```

```python
#add in proportion of gendered language
df['percen_gen'] = df['num_gen'] / df['num_queries']
```
